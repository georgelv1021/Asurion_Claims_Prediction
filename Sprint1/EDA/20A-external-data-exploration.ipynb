## Assessment and visualization

```python
#Loading external dataset
df1 = pd.read_excel("e1.xlsx", skiprows = 4)

df1 = df1.loc[:, ~df1.columns.str.contains('Unnamed: 0')]
df1 = df1.loc[:, ~df1.columns.str.contains('Unnamed: 3')]
df1.rename(columns={'Unnamed: 1':'Quarters', 'Unnamed: 2':'Number'}, inplace=True )

df1
```

```python
df2 = pd.read_excel("e2.xlsx", skiprows = 4)

df2 = df2.loc[:, ~df2.columns.str.contains('Unnamed: 0')]
df2 = df2.loc[:, ~df2.columns.str.contains('Unnamed: 10')]

df2.rename(columns={'Unnamed: 1':'Quarters', 'Unnamed: 2':'Number', 'Unnamed: 3':'Percentage'}, inplace=True )
df2
```

```python
df3 = pd.read_excel("e3.xlsx", skiprows = 4)
df3 = df3.loc[:, ~df3.columns.str.contains('Unnamed: 0')]
df3 = df3.loc[:, ~df3.columns.str.contains('Column25')]
df3 = df3.loc[:, ~df3.columns.str.contains('Unnamed: 27')]
df3
```

```python
# visualization on df1
q = df1['Quarters']
n = df1['Number']

plt.figure(figsize=(10,6))
plt.bar(q,n)
plt.xticks(rotation=90, fontsize = 'xx-small')
plt.title("Iphone shipment share against global phone shipment")
```

```python
# visualization on df2 
q = df2['Quarters']
apple = df2['Apple']
sam = df2['Samsung']

plt.figure(figsize=(10,6))
plt.bar(q,apple)
plt.bar(q,sam)
plt.xticks(rotation=90, fontsize = 'xx-small')
plt.title("Manufacture market share of iphone vs samsung")
```
